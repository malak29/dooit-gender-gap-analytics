```python
import pandas as pd

# Read the CSV files
df1 = pd.read_csv('Open_Payments_Data.csv')
df2 = pd.read_csv('research_2023.csv')

# Get the unique Covered_Recipient_Profile_ID from each DataFrame
ids_df1 = set(df1['Covered_Recipient_Profile_ID'])
ids_df2 = set(df2['Covered_Recipient_Profile_ID'])

# Find the common IDs
common_ids = ids_df1.intersection(ids_df2)

# Print the results
print(f"Number of unique IDs in Open_Payments_Data.csv: {len(ids_df1)}")
print(f"Number of unique IDs in research_2023.csv: {len(ids_df2)}")
print(f"Number of common IDs: {len(common_ids)}")

# If you want to see the common IDs (be cautious if there are many)
print("\nCommon IDs:")
print(common_ids)

# If you want to create a new DataFrame with only the common IDs
df_common = pd.concat([
    df1[df1['Covered_Recipient_Profile_ID'].isin(common_ids)],
    df2[df2['Covered_Recipient_Profile_ID'].isin(common_ids)]
])

# Save the common IDs to a new CSV file
df_common.to_csv('common_ids.csv', index=False)
print("\nCommon IDs saved to 'common_ids.csv'")
```

```python
import pandas as pd

# Read the CSV files
open_payments = pd.read_csv('Open_Payments_Data.csv')
research_2023 = pd.read_csv('research_2023.csv')

# Find common Covered_Recipient_Profile_IDs
common_ids = set(open_payments['Covered_Recipient_Profile_ID']) & set(research_2023['Covered_Recipient_Profile_ID'])

# List of companies
companies = [
    'Zimmer Biomet Holdings, Inc.', 'Medtronic, Inc.', 'Johnson & Johnson', 'Stryker Corporation', 
    'Abbott Laboratories', 'Boston Scientific Corporation', 'Intuitive Surgical, Inc.', 'Siemens Medical Solutions USA, Inc.', 
    'Becton, Dickinson and Company', 'Philips Electronics North America Corporation', 'Baxter Healthcare', '3M Company',
    'Alcon Inc.', 'Olympus', 'B. Braun', 'GE Healthcare', 'Fresenius Medical Care', 'Fujifilm Holdings Corporation', 'ABBVIE INC.', 'Arthrex, Inc.'
]

# Filter open_payments for common IDs and specified companies
open_payments_filtered = open_payments[
    (open_payments['Covered_Recipient_Profile_ID'].isin(common_ids)) &
    (open_payments['Submitting_Applicable_Manufacturer_or_Applicable_GPO_Name'].isin(companies))
]

# Columns to check
columns_to_check = ['Covered_Recipient_First_Name', 'Covered_Recipient_Middle_Name', 'Covered_Recipient_Last_Name']

# Check for non-null values in each column
for column in columns_to_check:
    if column in open_payments_filtered.columns:
        non_null_count = open_payments_filtered[column].notna().sum()
        total_count = len(open_payments_filtered)
        percentage = (non_null_count / total_count) * 100
        print(f"{column}:")
        print(f"  Non-null values: {non_null_count}")
        print(f"  Percentage: {percentage:.2f}%")
    else:
        print(f"{column} does not exist in the DataFrame")
    print()

# Display a few rows with non-null values in these columns
print("Sample rows with non-null values:")
sample_rows = open_payments_filtered[open_payments_filtered[columns_to_check].notna().any(axis=1)].head()
print(sample_rows[['Covered_Recipient_Profile_ID'] + columns_to_check])
```
